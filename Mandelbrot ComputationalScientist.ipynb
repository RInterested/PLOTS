```python
import numpy 
from numba import jit
import matplotlib.pyplot as plt
```

```python
@jit
def mandelbrot(Re, Im, max_iter):
    C = complex(Re,Im) # It will transform each value of the partitions of the x an y axis into a complex number.
    
    z= 0.0j # The Mandelbrot starts at zero (difference with Julia sets)

    for i in range(max_iter):
        z= z*z + C # It squares the previous complex number in the loop (starting at 0), and adds C.
        # Squaring a complex number (in polar coordinates) doubles the angle, and squares the module.
        if(z.real * z.real + z.imag * z.imag) >= 4: # If the length of the resulting z is greater than 2...
            return i  # The function returns the number of iterations it took for the function to blow up.

    return max_iter # Otherwise, it returns the number of iterations, which is the max, and means that it doesn't blow up.

n = 100
columns = n # The more columns, the more complex numbers will be evaluated. 200 is fast, and not too bad.
rows = n # Same here

# Baseline values to get a decent plot showing the main shape.
x1 = -2
x2 = 1
y1 = -1
y2 = 1


result = numpy.zeros([rows,columns]) 
for row_index, Re in enumerate(numpy.linspace(x1, x2, num=rows)):
    for column_index, Im in enumerate(numpy.linspace(y1, y2, num=columns)):
        result[row_index, column_index]= mandelbrot(Re, Im, 100)

plt.figure(dpi=100) 
plt.imshow(result.T,cmap='seismic',interpolation='bilinear', extent=[x1,x2,y1,y2]) 
plt.xlabel('Re') 
plt.ylabel('Im') 
plt.show()
```
